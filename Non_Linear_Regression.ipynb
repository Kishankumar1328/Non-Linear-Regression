### ***libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# ***linear***

```python
x=np.arange(-0.5,0.5,0.1)
y=2*(x)+3
y_hat=2*np.random.normal(size=x.size)
y_real=y+y_hat
plt.plot(x,y_real,"bo")
plt.plot(x,y,"-r")
plt.ylabel("dependent variable")
plt.xlabel("Independent variable")
```

### ***Quadratic***

# **𝑌=𝑋2**

```python
x=np.arange(13.0,16.0,0.1)
y=1*(x**3)+1*(x**2)+1*(x**1)+3
y_hat=20*np.random.normal(size=x.size)
kk=y_hat+y
plt.plot(x,kk,"bo")
plt.plot(x,y,color="hotpink",alpha=1)
plt.ylabel("Dependent variable")
plt.xlabel("Independent Variable")
```

## ***Exponential***

# **𝑌=𝑎+𝑏𝑐𝑋**


```python
x=np.arange(-13.0,29.0,0.2913)
y=np.exp(x)

plt.plot(x,y,color="silver",alpha=1)
plt.ylabel("Dependent variable")
plt.xlabel("Independent Variable")
```

# Logarithmic

# **𝑦=log(𝑥)**

```python
x=np.arange(-16.0,29.0,0.5)
y=np.log(x)
plt.ylabel("Dependent Variable")
plt.xlabel("Independent Variable")
plt.plot(x,y,color="gold",alpha=1)
```

### Sigmoidal/Logistic
# **𝑌=𝑎+𝑏1+𝑐(𝑋−𝑑)**


```python
x=np.arange(-13.0,29.0,1.6)
y=1+(-4/(1+pow(4,x-2)))
plt.ylabel("Dependent Variable")
plt.xlabel("Independent Variable")
plt.plot(x,y,color="yellowgreen",alpha=1)
```

```python
x=np.arange(13,29,0.4)#something different from other
y=1.0/(1.0+np.exp(-x))
plt.ylabel("Dependent Variable")
plt.xlabel("Independent Variable")
plt.plot(x,y,color="orange",alpha=1)
```
