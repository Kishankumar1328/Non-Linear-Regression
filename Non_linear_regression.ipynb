```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
def sigmoid(x,Beta_1,Beta_2):
  y=1/(1+np.exp(-Beta_1*(x-Beta_2)))
  return y
```

```python
beta_1=0.1
beta_2=1990.0

y_pred=sigmoid(x,beta_1,beta_2)

plt.plot(x,y_pred*1500000000000.)
plt.plot(x,y,"ro")
```
